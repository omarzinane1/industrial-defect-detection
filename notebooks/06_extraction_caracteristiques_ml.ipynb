# 06 - Extraction de caractéristiques pour le Machine Learning

## Objectif

Transformer les images en dataset tabulaire exploitable par un modèle de Machine Learning classique comme SVM.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.append(str(PROJECT_ROOT))

import pandas as pd

from src.features import build_features_dataframe, get_feature_columns
from src.utils import FEATURES_DIR, build_image_index, save_dataframe
```

## 1. Index des images

Chaque ligne représente une image avec son chemin, son split et son label.

```python
index_df = build_image_index()

if index_df.empty:
    raise FileNotFoundError("Aucune image trouvée dans data/raw/casting_data/.")

display(index_df.head())
print(f"Nombre d'images : {len(index_df)}")
```

## 2. Extraction des caractéristiques

Les caractéristiques utilisées sont : statistiques d'intensité, histogramme, texture GLCM, texture LBP et mesures de segmentation. Pour tester rapidement, vous pouvez mettre une valeur dans `MAX_IMAGES`.

```python
MAX_IMAGES = None  # exemple rapide: 1000
features_df = build_features_dataframe(index_df, max_images=MAX_IMAGES)

output_path = FEATURES_DIR / "casting_features.csv"
save_dataframe(features_df, output_path)

print(f"Fichier sauvegardé : {output_path}")
display(features_df.head())
```

## 3. Vérification du dataset tabulaire

On vérifie le nombre de caractéristiques numériques et la présence des labels.

```python
feature_columns = get_feature_columns(features_df)

print(f"Nombre de caractéristiques : {len(feature_columns)}")
print(feature_columns[:10])
display(features_df.groupby(["split", "label_name"]).size().reset_index(name="count"))
```

## Conclusion

Les images sont maintenant représentées sous forme de lignes numériques. Ce format est adapté à SVM et aux autres modèles classiques.
